```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 5,4
```

### Defining axes, ticks, and grids

```python
x = np.arange(1,10,1)
y = np.append(np.arange(1,5),0)
y = np.append(y, np.arange(4,0,-1))

fig = plt.figure()
ax = fig.add_axes([0.1,0.1,1,1])

ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([1,2,4,7,8,9])
ax.set_yticks([1,3,4])

ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.grid()

ax.plot(x,y)
```

### Generating multiple plots in one figure with subplots 

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.plot(x)
ax2.plot(x,y)
```

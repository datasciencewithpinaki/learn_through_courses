```python
z = range(1,5)
plt.pie(z)
plt.show()
```

### Customizing line styles

```python
cars.head()
```

```python
x = np.arange(len(cars.car_names))
y1 = cars.mpg
y2 = cars.qsec

plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.plot(x, y1, ls = '--', lw = 1) # line style; line width
plt.plot(x, y2, ds = 'steps', lw = 3) # drawing style
```

### Setting plot markers

```python
plt.plot(x, y1, marker = '1', mew = 20) # marker edge width
plt.plot(x, y2, marker = '+', mew = 15)
```

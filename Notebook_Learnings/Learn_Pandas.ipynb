```python
def generate_text(text_length):
    choice = np.random.randint(0,26,text_length)
    temp = []
    for i in choice:
        temp.append(string.ascii_uppercase[i])
    temp = ''.join(temp)
    return temp
```

```python
def myMapperFunction(col): 
    if col > 45:
        return 1
    else: 
        return 0
    
data_df['Customer_Age'][0:10].map(myMapperFunction)
myfilter = data_df['Customer_Age'].map(lambda x: x>45)
data_df['Age_Grtr_45'] = myfilter
data_df.head()
```

```python
def myApplyFunction(col, search_txt, replace_txt = 'new_'):
    if col.str.startswith(search_txt): 
        return replace_txt + col
    else: 
        return col
```

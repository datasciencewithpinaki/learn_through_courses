```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# multiple prints in a single cell
from IPython.core.interactiveshell import InteractiveShell 
InteractiveShell.ast_node_interactivity = 'all'
```

```python
data_df = pd.read_csv('BankChurners.csv')

data_df.shape
data_df.columns
```

```python
aggr_df = data_df.groupby('Education_Level', as_index = False)[['Customer_Age', 'Credit_Limit', 'Total_Trans_Amt']].agg('mean')

aggr_df.head()
aggr_df.columns
aggr_df.index
```

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
plt.figure(figsize=(7,5))

plt.bar(aggr_df['Education_Level'], aggr_df['Customer_Age'])
plt.ylim(40, 50)
plt.xlabel('Education Level')
plt.ylabel('Age')
plt.title('Plot between X axis vs Y axis')
plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.scatter(aggr_df['Education_Level'], aggr_df['Credit_Limit'], color = 'red', label = 'Credit Limit')

# plt.plot(aggr_df['Education_Level'], aggr_df['Credit_Limit'], 'bo', color = 'red', label = 'Credit Limit')
# plt.ylim(40, 50)
plt.xlabel('Education Level')
plt.ylabel('Credit Limit')
plt.title('Plot between X axis vs Y axis', fontsize = 20)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,10))

plt.bar(aggr_df['Education_Level'], aggr_df['Credit_Limit'], color = 'red', label = 'Credit Limit')
plt.bar(aggr_df['Education_Level'], aggr_df['Total_Trans_Amt'], color = 'blue', label = 'Total Transaction Amt')
plt.xlabel('Education Level')
plt.ylabel('Credit Limit')
plt.title('Plot between X axis vs Y axis')
plt.legend(loc = 'upper right')
plt.show()
```

```python
figure, axes = plt.subplots(1, 2, figsize=(12,5))

axes[0].bar(aggr_df['Education_Level'], aggr_df['Credit_Limit'], color = 'red')
axes[1].bar(aggr_df['Education_Level'], aggr_df['Total_Trans_Amt'], color = 'blue')

axes[0].set_title('Credit Limit')
axes[1].set_title('Total Transaction Amt')

for a in axes.flat:
    a.set(xlabel = 'Education Level', ylabel = 'In $')

plt.show()
```

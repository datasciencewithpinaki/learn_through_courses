```python
import numpy as np
import scipy.stats as stats

import matplotlib.pyplot as plt
import matplotlib.colors as pltcolors
```

```python
SIZE = 10_000
n = np.random.randn(SIZE)
n[0:4]
```

```python
# calculate bin size
def get_bin_size(data:np.array):
    n = len(data)
    h = 2*stats.iqr(data) * n**(-1/3)  # get bin width
    k = np.ceil((max(data) - min(data)) / h)
    return k
```

```python
bin_size = int(get_bin_size(n))
bin_size
```

```python
plt.hist(n, bins=40)
```

```python
fig, axs = plt.subplots(1,2, figsize=(12,4))
axs[0].hist(n, bins=10, color='b', alpha=0.7)
axs[1].hist(n, bin_size, color='g', alpha=0.3)

axs[0].set_title(f"binsize: {10}")
axs[1].set_title(f"binsize: {bin_size}")

plt.suptitle('histogram with different bins', fontsize=14)
plt.show();
```
